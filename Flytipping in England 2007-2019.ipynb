#### 1.Import Libraries

```python
import pandas as pd # for data manipulation and analysis
import seaborn as sns # to visualize random distributions
import matplotlib.pyplot as plt # for plotting
```

```python
df = pd.read_excel('flytipping land 2007-2019.xlsx', skipinitialspaces=True)
```

```python
pd.options.display.float_format = '{:,.0f}'.format
```

```python
df
```

#### 2. Plot Bar Chart Using Matplotlib

```python
df.plot(kind='bar', x='Year', y='Total Incidents')
```

```python
df1 = df
```

```python
df1_popTI = df1.pop('Total Incidents')
df1_popTI
```

```python
df1.head()
```

```python
df_melt = pd.melt(df1, id_vars=['Year'])
df_melt.head()
```

#### Introducing the function melt which is used to change the DataFrame format from wide to long i.e it turns columns into row data

```python
df_melt = df_melt.rename(columns={"variable": "Flytip_location", "value": "Number_of_incidents"})
df_melt.head()
```

#### 3. Plot Scatter diagram & Line Chart with Seaborn

```python
sns.catplot(x="Year", y="Number_of_incidents", hue="Flytip_location", data=df_melt, height=6, aspect=2).set(
           title = 'Number of Flytipping incidents and their location from 2007 to 2018/2019')
```

```python
plt.figure(figsize=(20, 8))
ax = sns.lineplot(x="Year", y="Number_of_incidents", hue="Flytip_location", data=df_melt).set(
    title = 'Line chart of number of flytipping incidents and their location from 2007 to 2018/19')
```
